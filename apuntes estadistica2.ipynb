* Ejemplo codigo

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Datos de ejemplo: satisfacción del cliente (y) en función de un predictor (x)
x = np.array([1, 2, 3, 4, 5])  # Variable independiente (por ejemplo, tiempo de espera)
y = np.array([3, 4, 5, 6, 8])  # Satisfacción del cliente

# Añadir constante para la regresión
X = sm.add_constant(x)

# Ajustar el modelo de regresión
modelo = sm.OLS(y, X).fit()

# Predicciones
y_pred = modelo.predict(X)

# Calcular Suma de Cuadrados Totales (SCT)
media_y = np.mean(y)
sct = np.sum((y - media_y) ** 2)

# Calcular Suma de Cuadrados de Regresión (SCR)
scr = np.sum((y_pred - media_y) ** 2)

# Calcular Suma de Cuadrados de Residuos (SCE)
sce = np.sum((y - y_pred) ** 2)

# Imprimir los resultados
print(f"Suma de Cuadrados Totales (SCT): {sct:.2f}")
print(f"Suma de Cuadrados de Regresión (SCR): {scr:.2f}")
print(f"Suma de Cuadrados de Residuos (SCE): {sce:.2f}")

# Crear gráficos
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Gráfico de los datos y la línea de regresión
axs[0].scatter(x, y, color='blue', label='Datos Reales')
axs[0].plot(x, y_pred, color='red', label='Línea de Regresión')
axs[0].set_title('Regresión Lineal')
axs[0].set_xlabel('Variable Independiente (X)')
axs[0].set_ylabel('Satisfacción del Cliente (Y)')
axs[0].legend()

# Gráfico de las sumas de cuadrados
axs[1].bar(['SCT', 'SCR', 'SCE'], [sct, scr, sce], color=['orange', 'green', 'purple'])
axs[1].set_title('Sumas de Cuadrados')
axs[1].set_ylabel('Valor')
axs[1].set_ylim(0, max(sct, scr, sce) + 5)  # Ajustar límite superior

# Mostrar la figura
plt.tight_layout()
plt.show()
```

* Al descomponer la variabilidad total en estas dos partes, podemos entender cuánto de la variabilidad se puede atribuir a la relación entre las variables (regresión) y cuánto se debe a errores aleatorios (residuos).
* Un SCResiduo alto en comparación con la SCTotal indica que el modelo es bueno para explicar la variabilidad en los datos.
* Un SCE bajo en relación con la SCT sugiere que el modelo se ajusta bien a los datos.
* Estos cálculos son fundamentales para calcular el coeficiente de determinación (R²), que es la proporción de la variabilidad total que se explica por el modelo. Se calcula como:
* R2: SCR/SCT
* Un R² alto indica que el modelo es efectivo para predecir la variable dependiente.
* En ANOVA y regresión, la relación entre las variables y la significancia estadística se evalúan mediante la comparación de estas sumas de cuadrados. Por ejemplo, se utilizan para calcular el valor F en ANOVA, lo que permite determinar si las diferencias entre grupos son estadísticamente significativas.
* Anova: La idea básica es ver si las diferencias observadas en las medias de los grupos son mayores que las que podríamos esperar por azar.
* Valor F: Si el valor F es alto, sugiere que la variabilidad entre las medias de los grupos es mayor que la variabilidad dentro de los grupos. Esto puede indicar que al menos uno de los grupos es significativamente diferente. Al calcular el valor F, lo comparamos con un valor crítico de F (obtenido de una tabla de distribución F) que depende del nivel de significancia (generalmente 0.05) y los grados de libertad de los grupos.

* Covarianza Positiva: Indica que cuando una variable aumenta, la otra tiende a aumentar también. Por ejemplo, si el ingreso y el gasto en consumo están positivamente relacionados, entonces una mayor ingreso se asocia con un mayor gasto.

* Covarianza Negativa: Indica que cuando una variable aumenta, la otra tiende a disminuir. Por ejemplo, si la temperatura y la cantidad de ropa que la gente usa están negativamente relacionadas, entonces a medida que la temperatura aumenta, la cantidad de ropa usada tiende a disminuir.

* Covarianza Cero: Sugiere que no hay una relación lineal entre las dos variables. Esto no significa que no haya relación en absoluto, ya que pueden estar relacionadas de otra manera (no lineal).

### 4. Estructura de datos:

* Nominal: datos se clasifican en categorías sin un orden específico. Ej: Género (masculino, femenino), colores (rojo, azul, verde), tipo de mascota (perro, gato, pez). No hay una relación cuantitativa entre las categorías.

* Ordinal: se clasifican en categorías que tienen un orden específico, pero la distancia entre las categorías no es cuantitativa. Clasificaciones (primer, segundo, tercer lugar), niveles de satisfacción (muy insatisfecho, insatisfecho, neutral, satisfecho, muy satisfecho). Se pueden clasificar y ordenar, pero no se puede medir la magnitud de la diferencia entre las categorías.

* Intervalos:  los datos se clasifican y ordenan, y las diferencias entre los valores son significativas. Se pueden realizar operaciones matemáticas como suma y resta. Ej: Temperatura en grados Celsius o Fahrenheit (donde 0 no significa "sin temperatura"), fechas en años.

* Relacion o razon: Es similar a la escala de intervalo, pero tiene un cero absoluto, lo que permite la comparación de magnitudes. Ej: Peso, altura, tiempo, ingresos (donde 0 significa "sin peso", "sin altura", "sin tiempo", "sin ingresos").

### 5. Rango Intercuantilico:
medida de dispersión que se utiliza en estadística para describir la variabilidad de un conjunto de datos. Específicamente, el RIC se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de un conjunto de datos.

* 1 Cuartil -> 25% de los datos es menor o igual a ese valor
* 2 cuartil -> 50 % de los datos es menor o igual a ese valor "mediana"
* 3 cuartil -> Valor total que el 75% de los datos es menor o igual a ese valor

Se calcula como:
* RIC=𝑄3−𝑄1

Este valor mide la dispersión de la mitad central de los datos, eliminando la influencia de valores atípicos o extremos.

El RIC es menos sensible a los valores atípicos que el rango total (la diferencia entre el valor máximo y mínimo), lo que lo convierte en una medida más confiable de la variabilidad.

### 6. Probabilidad

Se utiliza para medir la posibilidad de que ocurra un evento en relación con el total de posibles resultados.

* Evento: Un evento es un resultado o conjunto de resultados de un experimento aleatorio. Por ejemplo, lanzar un dado y obtener un número par (2, 4 o 6) es un evento.

* Espacio Muestral: El espacio muestral es el conjunto de todos los posibles resultados de un experimento. Por ejemplo, al lanzar un dado, el espacio muestral es {1, 2, 3, 4, 5, 6}.

* Probabilidad de un Evento: La probabilidad de un evento A se define como:

𝑃(𝐴) = Numero de resultados favorables/Numero total de resultados en el espacio muestral
 
La probabilidad de que un evento ocurra está siempre entre 0 (imposible) y 1 (seguro).

* Eventos Mutuamente Excluyentes: Dos eventos son mutuamente excluyentes si no pueden ocurrir al mismo tiempo. Por ejemplo, al lanzar un dado, obtener un 2 y un 5 en la misma tirada son eventos mutuamente excluyentes.

* Eventos Independientes: Dos eventos son independientes si la ocurrencia de uno no afecta la ocurrencia del otro. Por ejemplo, lanzar una moneda y lanzar un dado son eventos independientes.

* Reglas de Probabilidad:

Regla de adición: Para eventos mutuamente excluyentes 
P(A U 𝐵) = 𝑃(𝐴)+𝑃(𝐵)

Regla de multiplicación: Para eventos independientes 
P(A∩B)=P(A)×P(B)

* Probabilidad Condicional: La probabilidad de que ocurra un evento A dado que ocurrió un evento B se define como:


P(A∣B)= P(A∩B)/ P(B)
​
Esto se utiliza para evaluar la probabilidad de un evento en relación con otros eventos.

* Teorema de Bayes: Permite calcular la probabilidad de un evento basado en información previa. Es especialmente útil en situaciones donde se tiene información adicional.

### Aplicaciones de la Probabilidad

* Estadística: La probabilidad es fundamental para el análisis estadístico y la inferencia.
* Juegos de Azar: Se aplica para calcular las probabilidades de ganar en juegos como póker, ruleta, etc.
* Seguros: Las compañías de seguros utilizan la probabilidad para evaluar riesgos y establecer primas.
* Investigación Científica: En estudios experimentales, se utiliza para evaluar la significancia de los resultados.

**NOTA:** La probabilidad es una herramienta esencial para entender y cuantificar la incertidumbre en diversos campos. Permite hacer predicciones y tomar decisiones informadas basadas en la evaluación de eventos y sus posibilidades de ocurrencia.

#### Ejercicio 1: Probabilidad de un Evento

Calcula la probabilidad de obtener un número par al lanzar un dado de seis caras.

```python
# Definir el espacio muestral y los resultados favorables
total_resultados = 6  # Caras del dado
resultados_favorables = 3  # Números pares: 2, 4, 6

# Calcular la probabilidad
probabilidad_par = resultados_favorables / total_resultados
print(f"La probabilidad de obtener un número par al lanzar un dado es: {probabilidad_par:.2f}")
```

#### Ejercicio 2: Eventos Mutuamente Excluyentes

Calcula la probabilidad de obtener un 2 o un 5 al lanzar un dado.

```python
# Definir los eventos
evento_2 = 1  # Obtener un 2
evento_5 = 1  # Obtener un 5

# Calcular la probabilidad
probabilidad_2_o_5 = (evento_2 + evento_5) / total_resultados
print(f"La probabilidad de obtener un 2 o un 5 al lanzar un dado es: {probabilidad_2_o_5:.2f}")
```

#### Ejercicio 3: Eventos Independientes

Calcula la probabilidad de que al lanzar una moneda y un dado, obtengas cara en la moneda y un número par en el dado.

```python
# Definir las probabilidades
probabilidad_cara = 1 / 2  # Probabilidad de obtener cara
probabilidad_par = 3 / 6    # Probabilidad de obtener un número par

# Calcular la probabilidad de que ambos eventos ocurran
probabilidad_cara_y_par = probabilidad_cara * probabilidad_par
print(f"La probabilidad de obtener cara en la moneda y un número par en el dado es: {probabilidad_cara_y_par:.2f}")
```

#### Ejercicio 4: Probabilidad Condicional

Calcula la probabilidad de que sea viernes dado que es un día laboral (de lunes a viernes).

```python
# Definir el espacio muestral y eventos
dias_laborales = ['lunes', 'martes', 'miércoles', 'jueves', 'viernes']
dias_favorables = ['viernes']

# Probabilidad de que sea viernes dado que es día laboral
probabilidad_viernes = len(dias_favorables) / len(dias_laborales)
print(f"La probabilidad de que sea viernes dado que es un día laboral es: {probabilidad_viernes:.2f}")
```

#### Ejercicio 5: Teorema de Bayes

Supón que el 30% de los estudiantes de una clase estudian Matemáticas, y de esos, el 50% también estudian Física. Si un estudiante estudia Física, ¿cuál es la probabilidad de que también estudie Matemáticas?

```python
# Definir las probabilidades
P_M = 0.3  # Probabilidad de que un estudiante estudie Matemáticas
P_F_given_M = 0.5  # Probabilidad de estudiar Física dado que estudia Matemáticas
P_F = (P_M * P_F_given_M) + (0.7 * 0.4)  # Probabilidad total de estudiar Física (40% de los que no estudian Matemáticas)

# Aplicar el teorema de Bayes
P_M_given_F = (P_F_given_M * P_M) / P_F
print(f"La probabilidad de que un estudiante que estudia Física también estudie Matemáticas es: {P_M_given_F:.2f}")
P_F
```
